```python
from sklearn.model_selection import train_test_split
```

```python
df = dataset.drop(["PassengerId", "Name", "Cabin", "Ticket", "Embarked"], inplace=False, axis=1)
df["IsMale"] = (df["Sex"] == "male")
del df["Sex"]
df
```

```python
train, test = train_test_split(df, test_size=0.2)
```

```python
train_data = train[["Pclass", "Age", "SibSp", "Parch", "Fare", "IsMale"]]
train_labels = train["Survived"]

test_data = test[["Pclass", "Age", "SibSp", "Parch", "Fare", "IsMale"]]
test_labels = test["Survived"]
```

```python
train_data
```

```python
from catboost import CatBoostClassifier
```

```python
model = CatBoostClassifier(iterations=50,
                           learning_rate=1,
                           depth=3)
```

```python
model.fit(train_data, train_labels)
```

```python
preds_class = model.predict(test_data)
```

```python
preds_class
```

```python
for i in range(len(preds_class)):
    print(test_data.iloc[i])
    print('Label = ', test_labels.array[i], ', Prediced = ', preds_class[i], 'TRUE' if preds_class[i] == test_labels.array[i] else 'FALSE')
    print('-' * 80)
```

```python
dataset["Fare"].describe()
```

```python
from catboost import Pool

pool = Pool(train_data, train_labels, feature_names=list(train_data.columns))
```

```python
model.plot_tree(
    tree_idx=1,
    pool=pool)
```
